```python
%pylab inline
from kapteyn import maputils
from astropy.io import fits
#from matplotlib import pylab as plt

hdulist1 = fits.open('f336w_flc_cts.fits')
header1 = hdulist1[0].header
hdulist2 = fits.open('f438w_flc_cts.fits')
#header1 = hdulist1[0].header

fig = plt.figure(figsize=(8,8))
frame = fig.add_axes([0.15,0.15,0.8,0.8])
fx = maputils.FITSimage(externalheader=header1)
annim = fx.Annotatedimage(frame)
grat = annim.Graticule()
plt.set_cmap('Greys')
plt.imshow(hdulist1[0].data, vmin=0, vmax=50)
#plt.imshow(hdulist2[0].data, vmin=0, vmax=5)
#grat.setp_axislabel(fontstyle='italic')      # Apply to all
#grat.setp_axislabel("top", visible=True, xpos=0.0, ypos=1.0, rotation=180)
grat.setp_axislabel("bottom", label=r"$\mathrm{Right\ Ascension\ (2000)}$", fontsize=14)
grat.setp_axislabel("left", label=r"$\mathrm{Declination\ (2000)}$", fontsize=14)
annim.plot()
```

```python
% pylab inline
import numpy as np
from scipy.optimize import leastsq

f1 = open('meanmags.dat')
allbandmag = np.genfromtxt(f1)
f1.close()

f2 = open('feh.tsv')
metals = np.loadtxt(f2, skiprows=58)
f2.close()

alldata = []
for i in allbandmag:
    for j in metals:
        if i[0] == j[0] and j[7] == 1:
            alldata.append([i[0], i[1], i[2], i[3], i[4], i[5], i[6], i[7], i[8], i[9],
                           i[10], j[9], j[1], j[2]])
        if i[0] == j[0] and j[7] == 0:
            alldata.append([i[0], i[1], i[2], i[3], i[4], i[5], i[6], i[7], i[8], i[9],
                           i[10], j[9] + 0.128, j[1], j[2]])
```

```python
def pl_fit(params, data):
    return params[0]*np.log10(data[2]) + params[1] + params[2]*(data[3] + 1.6)

def fitfunc(dmod, params, data):
    return (pl_fit(params, data) + dmod) - data[0]

# print alldata[0]

# Mv =  0.5229 - 0.0000 logP + 0.277(Fe/H+1.6)
# Mi = -0.2198 - 1.1318 logP + 0.2053(Fe/H+1.6)
# Mj = -0.7799 - 1.7734 logP + 0.1899(Fe/H+1.6)
# Mh = -1.1500 - 2.3134 logP + 0.1780(Fe/H+1.6)
# Mk = -1.1844 - 2.3529 logP + 0.1746(Fe/H+1.6)

vparams = [-1.0, 0.5229, 0.277]
jparams = [-1.7734, -0.7799, 0.1899]
hparams = [-2.3134, -1.1500, 0.1780]
kparams = [-2.3529, -1.1844, 0.1746]

ad_zip = zip(*alldata)

per = array(ad_zip[11])
feh = array(ad_zip[12])

def filter_nans(data_raw):
    data = zip(*data_raw)
    data_filtered = []
    for i in data:
        if str(i[0]) != 'nan' and i[0] > 11.:
            data_filtered.append(i)
    d = zip(*data_filtered)
    return [array(d[0]), array(d[1]), array(d[2]), array(d[3]), array(d[4])]

vdata_raw = [array(ad_zip[3]), array(ad_zip[4]), per, feh, array(ad_zip[13])]
jdata_raw = [array(ad_zip[5]), array(ad_zip[6]), per, feh, array(ad_zip[13])]
hdata_raw = [array(ad_zip[7]), array(ad_zip[8]), per, feh, array(ad_zip[13])]
kdata_raw = [array(ad_zip[9]), array(ad_zip[10]), per, feh, array(ad_zip[13])]

vdata = filter_nans(vdata_raw)
jdata = filter_nans(jdata_raw)
hdata = filter_nans(hdata_raw)
kdata = filter_nans(kdata_raw)

dmod=14

v_out = leastsq(fitfunc, dmod, args=(vparams, vdata), full_output=True)
j_out = leastsq(fitfunc, dmod, args=(jparams, jdata), full_output=True)
h_out = leastsq(fitfunc, dmod, args=(hparams, hdata), full_output=True)
k_out = leastsq(fitfunc, dmod, args=(kparams, kdata), full_output=True)

v_dmod = float(v_out[0])
v_err = float(sqrt(v_out[1]))
j_dmod = float(j_out[0])
j_err = float(sqrt(j_out[1]))
h_dmod = float(h_out[0])
h_err = float(sqrt(h_out[1]))
k_dmod = float(k_out[0])
k_err = float(sqrt(k_out[1]))

print "V:", v_dmod, "+/-", v_err
print "J:", j_dmod, "+/-", j_err
print "H:", h_dmod, "+/-", h_err
print "K:", k_dmod, "+/-", k_err
```

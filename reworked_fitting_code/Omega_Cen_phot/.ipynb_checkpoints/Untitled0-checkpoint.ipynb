```python
#2 = Kaluzny Vband   3 = Kaluzny Bband -1 = Fourstar H    -2 = Fourstar Ks  -5 = Fourstar J
fuckindata = []
import numpy as np
import glob
for filename in glob.glob('*.txt'):
    k_v = []
    k_b = []
    f_h = []
    f_k = []
    f_j = []
    f = open(filename)
    data = np.genfromtxt(f)
    for i in data:
        if i[3] == 2:
            k_v.append(i)
        if i[3] == 3:
            k_b.append(i)
        if i[3] == -1:
            f_h.append(i)
        if i[3] == -2:
            f_k.append(i)
        if i[3] == -5:
            f_j.append(i)
    #print len(k_v), len(k_b), len(f_h), len(f_k), len(f_j)
    #convert mag to flux, average, back again
    flux_v = []
    flux_b = []
    flux_h = []
    flux_k = []
    flux_j = []
    err_v = []
    err_b = []
    err_h = []
    err_k = []
    err_j = []
    for i in k_v:
        flux = 10**(i[1]/-2.5)
        flux_v.append(flux)
        err_v.append(i[2]**2)
    for i in k_b:
        flux = 10**(i[1]/-2.5)
        flux_b.append(flux)
        err_b.append(i[2]**2)
    for i in f_h:
        flux = 10**(i[1]/-2.5)
        flux_h.append(flux)
        err_h.append(i[2]**2)
    for i in f_j:
        flux = 10**(i[1]/-2.5)
        flux_j.append(flux)
        err_j.append(i[2]**2)
    for i in f_k:
        flux = 10**(i[1]/-2.5)
        flux_k.append(flux)
        err_k.append(i[2]**2)
    mean_mag_v = np.round(-2.5*np.log10(np.mean(flux_v)), decimals=7)
    mean_mag_b = np.round(-2.5*np.log10(np.mean(flux_b)), decimals=7)
    mean_mag_h = np.round(-2.5*np.log10(np.mean(flux_h)), decimals=7)
    mean_mag_j = np.round(-2.5*np.log10(np.mean(flux_j)), decimals=7)
    mean_mag_k = np.round(-2.5*np.log10(np.mean(flux_k)), decimals=7)
    mean_err_v = np.round(np.sqrt(np.sum(err_v)), decimals=7)
    mean_err_b = np.round(np.sqrt(np.sum(err_b)), decimals=7)
    mean_err_h = np.round(np.sqrt(np.sum(err_h)), decimals=7)
    mean_err_j = np.round(np.sqrt(np.sum(err_j)), decimals=7)
    mean_err_k = np.round(np.sqrt(np.sum(err_k)), decimals=7)
    fuckindata.append([filename[:3], mean_mag_b, mean_err_b, mean_mag_v, mean_err_v,
                       mean_mag_j, mean_err_j, mean_mag_h, mean_err_h, mean_mag_k, mean_err_k])
```

```python
f = open('meanmags.dat', 'w')
for i in fuckindata:
    f.write('  '.join(map(str, i)) + '\n')
f.close()
```
